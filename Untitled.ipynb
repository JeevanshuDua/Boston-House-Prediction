```python
from sklearn import linear_model
import pandas as pd
from sklearn.datasets import load_boston
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
boston = load_boston()
boston
```

```python
boston.target
```

```python
X = pd.DataFrame(boston.data, columns = boston.feature_names)
X.head()
```

```python
y = pd.DataFrame(boston.target, columns = ["MEDV"])
y.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(y.MEDV, bins=30)
plt.show()
```

```python
plt.scatter(y.index, y.MEDV)
```

```python
lm = linear_model.LinearRegression()
```

```python
lm.fit(X,y)
```

```python
pred = lm.predict(X)
pred = pd.DataFrame(pred)
pred.head()
```

```python
lm.score(X,y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.scatter(y.index, y, label='Actual', color='b')
plt.plot(pred, label='Predicted', color='r')
```

## CHi-Square

```python
import scipy.stats as stats
from scipy.stats import chi2_contingency

Significance = pd.Series()

for i in X.columns:
    dfObserved = pd.concat([X[i],y], axis=0)
    chi2, p, dof, expected = stats.chi2_contingency(dfObserved.values)
    Significance[i] = p

Significance.sort_values()
```

## VIF (Variance Inflation Factor)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

X = add_constant(X)

pd.Series([variance_inflation_factor(X.values, i) 
               for i in range(X.shape[1])], 
              index=X.columns)
```
